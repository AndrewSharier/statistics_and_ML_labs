# **Задача №1**

1. Определить (с обоснованием), зависимы или независимы следующие события:

(a) Несовместные события;

(b) События, образующие σ-алгебру Σ в пространстве (Ω, Σ, P);

(c) События, имеющие одинаковую вероятность?

Независимость событий определяется следующим выражением (вероятность их общего наступления равна произведению их вероятностей):

P(A ∩ B) = P(A) * P(B)

# **Решение:**

а) несовместные события - события, которые не могут произойти одновременно, соответственно вероятность пересечения таких событий равна пустому множеству:
P(A ∩ B) = P(∅)= 0

Следовательно, несовместные события могут быть независимыми, если вероятность наступления одного из них равно 0 (P(A) = 0, например), то есть событие является невозможным. Соответственно, дать однозначный ответ в этом пунтке возможно, если известны вероятности наступления обоих событий или известна природа этих событий.

b) факт принадлежности событий к σ-алгебре Σ в пространстве (Ω, Σ, P) не говорит о зависимости или независимости этих событий.

Например, если σ-алгебра порождена пустым множеством и событием А (то есть Σ = {∅, А, -А, Ω), то события А и -А являются несовместными, а значит, зависимыми (из доказанного в пункте a и при условии, что P(A)!=0).

Другой пример: рассмотрим σ-алгебру, порожденную множеством элементарных исходов подбрасывания монеты 2 раза. σ-алгебра имеет вид Σ = (∅, ОО, ОР, РО, РР, Ω), где О - соответственно выпадение орла, а Р - выпадение решки.

Рассмотрим события "В первый раз выпал орел" {OO,OP} (P({OO,OP})=0.5) и "Во второй раз выпала решка" {OP,PP} (P({OP,PP})=0.5).

Событием, соответсвующим пересечению этих собтий, является событие {OP} (P({OP}=0.25). В то же время P({OO,OP}) * P({OP,PP}) = 0.5 * 0.5 = 0.25, то есть два рассматриваемых события являются независимыми.

с) события, имеющие одинаковую вероятность. Это не говорит о зависимости или независимости событий, и события могут быть равновероятно как завсимыми, так и независимыми в случае, если их вероятности равны.

Пример независимых событий: из пункта b про подбрасывание монеты 2 раза.

Пример зависимых событий: подбрасывание монеты 1 раз. События "Выпал орел" и "Выпала решка", которые составляют множество элементарных исходов, являются несовместными и, следовательно, зависимыми (из доказанного в пункте a; P('O') = P('P') = 0.5

# **Задача №2**

Опыт заключается в независимом подбрасывании двух симметричных
монет. Рассматриваются следующие события:

• A — появление герба на первой монете;

• B — появление решки на первой монете;

• C — появление герба на второй монете;

• D — появление решки на второй монете;

• E — появление хотя бы одного герба;

• F — появление хотя бы одной решки;

• G — появление одного герба и одной решки;

• H — непоявление ни одного герба;

• K — появление двух гербов.

Определить, каким событиям этого списка равносильны следующие события:

(a) A + C = ?

(b) AC = ?

(c) EF = ?

(d) G + E = ?

(e) GE = ?

(f) BD = ?

(g) E + K = ?

# **Решение**

(a) A + C = "появление герба на первой монете" + "появление герба на второй монете" = "появление двух гербов" (*событие K*)

(b) AC = "появление герба на первой монете" * "появление герба на второй монете" = "появление двух гербов" (*событие K*)

(c) EF = "появление хотя бы одного герба" * "появление хотя бы одной решки" = "появление одного герба и одной решки" (*событие G*)

(d) G + E = "появление одного герба и одной решки" + "появление хотя бы одного герба" = "появление одного герба и одной решки" (*событие G*)

(e) GE = "появление одного герба и одной решки" * "появление хотя бы одного герба" = "появление одного герба и одной решки" (*событие G*)

(f) BD = "появление решки на первой монете" * "появление решки на второй монете" = "непоявление ни одного герба" (*событие H*)

(g) E + K = "появление хотя бы одного герба" + "появление двух гербов" = "появление двух гербов" (*событие K*)

# **Задача №4**

Два парохода должны подойти к одному и тому же причалу независимо друг от друга и равновозможно в течение суток. Определить вероятность того, что одному из них придется ожидать освобождения причала, если время стоянки первого парохода — 1 час, а второго — 2 часа.

# **Решение**

Возможные исходы:

1. пароходу А придется ожидать освобождения причала пароходом Б (то есть x ∈ [y,y+2] - пароход А пришел, пока Б стоит);

2. пароходу Б придется ожидать освобождения причала пароходом А (то есть y ∈ [x,x+1] - пароход Б пришел, пока А стоит);

3. ни одному из пароходов не придется ожидать освобождения причала.

Для нахождения вероятности будем использовать геометрическое опредление вероятности. Все множество элементарных исходов представляет собой квадрат 24х24 (x - ось, соответствующая времени прибытия парохода А; y - ось, соответствующая времени прибытия парохода Б).

S = 24 x 24 = 576

Искомую вероятность можно вычислить как отношение площадей фигуры, описывающей стоянку одного из пароходов, к площади квадрата S.

Площадь фигуры, связанная со стоянкой одного из пароходов, ограничена двумя прямыми, которые описываются следующими выражениями:

x <= y <= x + 1

y <= x <= y + 2, откуда y => x - 2

Таким образом, прямые *y = x - 2*  и *y = x + 1* разделяют исходный квадрат множества элементарных событий на 3 фигуры. Фигура выше прямой *y = x + 1* - треугольник с катетами 23 и вершинами в точках (0,1), (0,24), (23,24) (= прибытие парохода Б позже парохода А без ожидания стоянки пароходом Б). Фигура ниже прямой *y = x - 2* - треугольник с катетами 22 и вершинами в точках (2,0), (24,0), (24,22) (= прибытие парохода Б раньше парохода А без ожидания стоянки пароходом А).

S(Б без стоянки) = 1/2 * 23 * 23 = 529/2

S(А без стоянки) = 1/2 * 22 * 22 = 484/2

S(ож. стоянки) = 576 - 529/2 - 484/2 = 139/2

P(ож. стоянки) = (139/2) / 576 = 139 / 1152 = 0.121

# **Задача №5**

Самолет, по которому ведется стрельба, состоит из трех различных по
уязвимости частей:

(a) Кабина летчика и двигатель

(b) Топливные баки

(c) Планер

Для поражения самолета достаточно либо одного попадания в первую часть, либо двух попаданий во вторую, либо трех в третью. При попадании в самолет одного снаряда, снаряд с вероятностью p1 попадает в первую часть, с вероятностью p2 — во вторую, с вероятностью p3 — в третью. Попавшие снаряды распределяются по частям независимо друг от друга.

Известно, что в самолет попало m снарядов. Найти условную вероятность P(A|m) события A – «Самолет поражен» – при m = 1, 2, 3, 4.

# **Решение**

При m = 1, событие "самолет поражен" может наступить только при попадании в первую часть (a): P(A|m = 1) = p1

При m = 2, событие "самолет поражен" наступает при одном из попаданий в первую часть, либо при обоих попаданиях во вторую часть: (1,1,0), (1,0,1), (0,2,0)

p(1,1,0) = 2!/(1!1!0!) * p1 * p2 = 2p1p2

p(1,0,1) = 2!/(1!0!1!) * p1 * p3 = 2p1p3

p(0,2,0) = 2!/(0!2!0!) * p2^2 = p2^2

Итого, P(A|m = 2) = p2^2 + 2p1p2 + 2p1p3

При m = 3, проще рассматривать событие "самолет НЕпоражен", которое может наступить при следующем исходе: (0,1,2)

p(0,1,2) = 3!/(0!1!2!) * p2 * p3^2 = 3p2(p3^2)

Итого, P(A|m = 3) = 1 - p(0,1,2) = 1 - 3p2(p3^2)

При m = 4, событие "самолет поражен" наступает при любом исходе, поэтому P(A|m = 4) = 1

# **1.2 Случайный вектор и числовые характеристики**

# **Задача №1**

Пусть

fξ(x, y) = e^(−2|y|) / π(1 + x^2)

Является ли данная функция плотностью распределения случайного
вектора?

# **Решение**

Для проверки того, является ли указанная функция плотностью распределения вектора (x, y), необходимо проверить несколько условий:

1) условие неотрицательности (fξ(x, y) => 0 при всех (x, y))

2) условие нормированности (интеграл по всей плоскости (x, y) = 1)


---


1. Данная функция является сложной функцией, состоящей из двух функций g(y) = e^(−2|y|) и h(x) = 1 / (1 + x^2) и константы π

Функция g(y) = e^(−2|y|) является экспоненциальной и всегда положительна.

Функция h(x) = 1 / (1 + x^2) является смесью параболы и гиперболы, но при любом x она положительна, так как знаменатель функции всегда положителен и не равен 0.

Следовательно, функция fξ(x, y) положительна при всех (x, y)

2. Так как функция fξ(x, y) состоит из двух функций, то и двойной интеграл от функции fξ(x, y) может быть расписан как произведение интеграла от функции g(y) = e^(−2|y|) по dy на интеграл от функции h(x) = 1 / (1 + x^2) по dx

Интеграл функции h(x) = 1 / (1 + x^2) по dx = arctan(x). Интегрируя по всей плоскости от −∞ до +∞ получаем, что arctan(x) = π/2 - (-π/2) = π

Интеграл функции g(y) = e^(−2|y|) может быть представлен в виде суммы интегралов: интеграл от функции e^(−2*(-y)) = e^(2y) (интегриуремой по плоскости от −∞ до 0) плюс интеграл от функции e^(−2y) (интегриуремой по плоскости от 0 до +∞).


​Интеграл от функции e^(2y), интегрируемый в указанной области, равен 1/2 * e^(2y) = 1/2 * (1-0) = 1/2

Интеграл от функции e^(−2y), интегрируемый в указанной области, равен -1/2 * e^(−2y) = -1/2 * (0-1) = 1/2


---


Таким образом, интеграл от функции g(y) = e^(−2|y|), взятый по плоскости от −∞ до +∞, равен 1/2 + 1/2 = 1.

А интеграл от заданной функции, интегрируемой по всей плоскости R2, равен = 1 * π * 1/π = 1, то есть условие нормирвоанности выполняется.

Следовательно, указанная функция явлется плотностью распределения случайного вектора (x, y)
